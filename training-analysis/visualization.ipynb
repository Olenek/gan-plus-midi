```python
import plotly.graph_objects as go
import pandas as pd
```

```python
gen = pd.read_csv('gen-loss.csv')
gen.columns = ['wall', 'step', 'loss']

disc = pd.read_csv('disc-loss.csv')
disc.columns = ['wall', 'step', 'loss']
```

```python
df = gen.merge(disc, on='step', suffixes=('_gen', '_disc'))
```

```python
df
```

```python
df['loss_disc_ma5'] = df.loss_disc.rolling(5).mean()
df['loss_gen_ma5'] = df.loss_gen.rolling(5).mean()
```

```python
trace1 = go.Scatter(
    x=df['step'],
    y=df['loss_gen'],
    name='Generator',
)

trace2 = go.Scatter(
    x=df['step'],
    y=df['loss_disc'],
    name='Discriminator',    
)
```

```python
from plotly.subplots import make_subplots

fig = go.Figure(data=[trace1, trace2])
fig.update_layout(
    title='Loss over Epochs of Training',
    xaxis_title="Epoch",
    yaxis_title="Wasserstein Loss",
    xaxis_range=[0, 2000]
)
fig.show()
```
